```python
import pandas as pd
import pickle
import numpy as np
import pathlib
```

```python
energy_files = list(pathlib.Path("QAOA_Results_Backup 2").rglob("*.npz"))
```

```python
results  = {}
with open(energy_files[0], 'rb') as f:
    saved_results = np.load(f, allow_pickle=True)

    keys = list(saved_results.keys())
    for key in keys:
        results[key] = saved_results[key]
```
