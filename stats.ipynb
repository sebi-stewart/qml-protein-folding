```python
from statsmodels.stats.power import TTestIndPower

# Parameters to yield ~30
analysis = TTestIndPower()
effect_size = 0.66  # Assumed biological significance threshold
alpha = 0.05
power = 0.80

n = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=1.0)
print(f"Required samples per group: {round(n)}")
```
