```python
import numpy as np
import pandas as pd

def create_energy_matrix(one_body, two_body):
    """
    Maps one-body and two-body energies to a matrix suitable for surface plotting.

    Args:
        one_body: {residue_id: {rotamer_id: energy}}
        two_body: {(residue_i, residue_j): {(rotamer_i, rotamer_j): energy}}

    Returns:
        matrix: 2D numpy array where indices represent (residue, rotamer) combinations
        index_to_key: mapping from matrix index back to (residue_id, rotamer_id)
    """

    # Step 1: Create a mapping of all (residue, rotamer) combinations to flat indices
    flat_indices = {}
    index_to_key = {}
    current_idx = 0

    # Sort residues for consistent ordering
    for res_id in sorted(one_body.keys()):
        num_rotamers = len(one_body[res_id])
        for rot_id in range(num_rotamers):
            flat_indices[(res_id, rot_id)] = current_idx
            index_to_key[current_idx] = (res_id, rot_id)
            current_idx += 1

    total_elements = current_idx

    # Step 2: Create the energy matrix
    energy_matrix = np.full((total_elements, total_elements), np.nan)

    # Step 3: Fill in one-body energies (diagonal)
    for res_id, rotamers in one_body.items():
        for rot_id, energy in rotamers.items():
            idx = flat_indices[(res_id, rot_id)]
            energy_matrix[idx, idx] = energy

    # Step 4: Fill in two-body energies (off-diagonal)
    for (res_i, res_j), interactions in two_body.items():
        for (rot_i, rot_j), energy in interactions.items():
            idx_i = flat_indices[(res_i, rot_i)]
            idx_j = flat_indices[(res_j, rot_j)]

            # Store symmetrically (since it's an interaction)
            energy_matrix[idx_i, idx_j] = energy
            energy_matrix[idx_j, idx_i] = energy

    return energy_matrix, index_to_key, flat_indices
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

def plot_energy_landscape(energy_matrix, index_to_key):
    """
    Creates a 3D surface plot of the energy landscape.
    NaN values (invalid interactions) are handled by the plot.
    """
    # Create coordinate grids
    x = np.arange(energy_matrix.shape[1])
    y = np.arange(energy_matrix.shape[0])
    X, Y = np.meshgrid(x, y)

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Plot surface (NaN values are automatically excluded)
    surf = ax.plot_surface(X, Y, energy_matrix, cmap='viridis',
                           alpha=0.9, rstride=1, cstride=1)

    ax.set_xlabel('(Residue, Rotamer) Index - j')
    ax.set_ylabel('(Residue, Rotamer) Index - i')
    ax.set_zlabel('Energy')
    ax.set_title('Protein Energy Landscape')

    fig.colorbar(surf, ax=ax, label='Energy (kcal/mol)')

    return fig, ax
```

```python
def create_interaction_only_matrix(one_body, two_body):
    """
    Creates a matrix of only two-body interactions, useful if one-body
    energies are treated separately.
    """
    flat_indices = {}
    current_idx = 0

    for res_id in sorted(one_body.keys()):
        num_rotamers = len(one_body[res_id])
        for rot_id in range(num_rotamers):
            flat_indices[(res_id, rot_id)] = current_idx
            current_idx += 1

    total_elements = current_idx
    interaction_matrix = np.full((total_elements, total_elements), np.nan)

    # Only fill two-body terms
    for (res_i, res_j), interactions in two_body.items():
        for (rot_i, rot_j), energy in interactions.items():
            idx_i = flat_indices[(res_i, rot_i)]
            idx_j = flat_indices[(res_j, rot_j)]
            interaction_matrix[idx_i, idx_j] = energy
            interaction_matrix[idx_j, idx_i] = energy

    return interaction_matrix, flat_indices
```

```python
def plot_energy_heatmap(energy_matrix, index_to_key):
    """
    Creates a 2D heatmap of the energy landscape.
    NaN values are nicely masked (shown as white/blank).
    """
    fig, ax = plt.subplots(figsize=(14, 12))

    # Create labels
    labels = [f"R{k[0]}_Rot{k[1]}" for k in [index_to_key[i] for i in range(len(index_to_key))]]

    # Use masked array to handle NaN values
    masked_matrix = np.ma.masked_invalid(energy_matrix)

    im = ax.imshow(masked_matrix, cmap='viridis', aspect='auto', interpolation='nearest')

    ax.set_xticks(range(len(labels)))
    ax.set_yticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=8)
    ax.set_yticklabels(labels, fontsize=8)

    ax.set_xlabel('(Residue, Rotamer) Index')
    ax.set_ylabel('(Residue, Rotamer) Index')
    ax.set_title('Protein Energy Landscape - Heatmap')

    fig.colorbar(im, ax=ax, label='Energy (kcal/mol)')
    plt.tight_layout()

    return fig, ax

from scipy.interpolate import griddata

def plot_energy_landscape_triangulated(energy_matrix, index_to_key):
    """
    Creates a triangulated 3D surface from valid energy points.
    This avoids NaN-induced gaps.
    """
    # Extract valid (non-NaN) points
    valid_points = []
    valid_energies = []

    for i in range(energy_matrix.shape[0]):
        for j in range(energy_matrix.shape[1]):
            if not np.isnan(energy_matrix[i, j]):
                valid_points.append([i, j])
                valid_energies.append(energy_matrix[i, j])

    valid_points = np.array(valid_points)
    valid_energies = np.array(valid_energies)

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Use plot_trisurf for triangulated surface
    surf = ax.plot_trisurf(valid_points[:, 0], valid_points[:, 1], valid_energies,
                           cmap='viridis', alpha=0.9, edgecolor='none')

    ax.set_xlabel('(Residue, Rotamer) Index - j')
    ax.set_ylabel('(Residue, Rotamer) Index - i')
    ax.set_zlabel('Energy')
    ax.set_title('Protein Energy Landscape - Triangulated Surface')

    fig.colorbar(surf, ax=ax, label='Energy (kcal/mol)')

    return fig, ax

def create_compact_interaction_matrix(one_body, two_body):
    """
    Creates a compact matrix excluding self-interaction blocks.
    Results in a smaller matrix without NaN values.
    """
    flat_indices = {}
    index_to_key = {}
    current_idx = 0

    for res_id in sorted(one_body.keys()):
        num_rotamers = len(one_body[res_id])
        for rot_id in range(num_rotamers):
            flat_indices[(res_id, rot_id)] = current_idx
            index_to_key[current_idx] = (res_id, rot_id)
            current_idx += 1

    total_elements = current_idx
    compact_matrix = np.zeros((total_elements, total_elements))

    # Fill in one-body energies on diagonal
    for res_id, rotamers in one_body.items():
        for rot_id, energy in rotamers.items():
            idx = flat_indices[(res_id, rot_id)]
            compact_matrix[idx, idx] = energy

    # Fill in two-body energies (off-diagonal only)
    for (res_i, res_j), interactions in two_body.items():
        for (rot_i, rot_j), energy in interactions.items():
            idx_i = flat_indices[(res_i, rot_i)]
            idx_j = flat_indices[(res_j, rot_j)]

            # Only store if different residues
            if res_i != res_j:
                compact_matrix[idx_i, idx_j] = energy
                compact_matrix[idx_j, idx_i] = energy

    return compact_matrix, index_to_key

def plot_energy_scatter(energy_matrix, index_to_key):
    """
    Scatter plot with energy values color-coded.
    Cleaner alternative when surface has too many gaps.
    """
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Extract valid points
    x, y, z = [], [], []
    for i in range(energy_matrix.shape[0]):
        for j in range(energy_matrix.shape[1]):
            if not np.isnan(energy_matrix[i, j]):
                x.append(i)
                y.append(j)
                z.append(energy_matrix[i, j])

    scatter = ax.scatter(x, y, z, c=z, cmap='viridis', s=50, alpha=0.6)

    ax.set_xlabel('(Residue, Rotamer) Index - i')
    ax.set_ylabel('(Residue, Rotamer) Index - j')
    ax.set_zlabel('Energy')
    ax.set_title('Protein Energy Landscape - Scatter Plot')

    fig.colorbar(scatter, ax=ax, label='Energy (kcal/mol)')



    return fig, ax
```

```python
import matplotlib.colors as colors

def shift_and_log_data(energy_matrix):
    """
    Shifts energy data so all values are positive, then applies log scaling.
    Handles NaN values appropriately.

    Args:
        energy_matrix: 2D array with possibly negative values and NaNs

    Returns:
        log_matrix: Log-scaled matrix with all positive values
        shift_amount: Amount added to make data positive
    """
    # Find minimum value (ignoring NaN)
    min_val = np.nanmin(energy_matrix)

    # Shift so minimum is slightly above 0
    # Add 1 to ensure log(1) = 0 for the minimum value
    shift_amount = abs(min_val) + 1 if min_val < 0 else 1

    # Create shifted matrix
    shifted_matrix = energy_matrix + shift_amount

    # Apply log scale (NaN values remain NaN)
    log_matrix = np.log10(shifted_matrix)

    print(f"Data shifted by: {shift_amount}")
    print(f"Original range: [{np.nanmin(energy_matrix):.4f}, {np.nanmax(energy_matrix):.4f}]")
    print(f"Shifted range: [{np.nanmin(shifted_matrix):.4f}, {np.nanmax(shifted_matrix):.4f}]")
    print(f"Log range: [{np.nanmin(log_matrix):.4f}, {np.nanmax(log_matrix):.4f}]")

    return log_matrix, shift_amount

def plot_energy_heatmap_log(energy_matrix, index_to_key):
    """
    Creates a 2D heatmap of the energy landscape with log scaling.
    """
    # Get log-scaled data
    log_matrix, shift_amount = shift_and_log_data(energy_matrix)

    fig, ax = plt.subplots(figsize=(14, 12))

    # Create labels
    labels = [f"R{k[0]}_Rot{k[1]}" for k in [index_to_key[i] for i in range(len(index_to_key))]]

    # Use masked array to handle NaN values
    masked_matrix = np.ma.masked_invalid(log_matrix)

    im = ax.imshow(masked_matrix, cmap='viridis', aspect='auto', interpolation='nearest')

    ax.set_xticks(range(len(labels)))
    ax.set_yticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=8)
    ax.set_yticklabels(labels, fontsize=8)

    ax.set_xlabel('(Residue, Rotamer) Index')
    ax.set_ylabel('(Residue, Rotamer) Index')
    ax.set_title(f'Protein Energy Landscape - Heatmap (Log Scale, shifted by {shift_amount:.2f})')

    cbar = fig.colorbar(im, ax=ax, label='log₁₀(Energy + shift)')
    plt.tight_layout()

    return fig, ax

def plot_energy_landscape_triangulated_log(energy_matrix, index_to_key):
    """
    Creates a triangulated 3D surface with log scaling.
    """
    # Get log-scaled data
    log_matrix, shift_amount = shift_and_log_data(energy_matrix)

    # Extract valid (non-NaN) points
    valid_points = []
    valid_energies = []

    for i in range(log_matrix.shape[0]):
        for j in range(log_matrix.shape[1]):
            if not np.isnan(log_matrix[i, j]):
                valid_points.append([i, j])
                valid_energies.append(log_matrix[i, j])

    valid_points = np.array(valid_points)
    valid_energies = np.array(valid_energies)

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Use plot_trisurf for triangulated surface
    surf = ax.plot_trisurf(valid_points[:, 0], valid_points[:, 1], valid_energies,
                           cmap='viridis', alpha=0.9, edgecolor='none')

    ax.set_xlabel('(Residue, Rotamer) Index - j')
    ax.set_ylabel('(Residue, Rotamer) Index - i')
    ax.set_zlabel('log₁₀(Energy + shift)')
    ax.set_title(f'Protein Energy Landscape - Triangulated Surface (Log Scale, shifted by {shift_amount:.2f})')

    fig.colorbar(surf, ax=ax, label='log₁₀(Energy + shift)')

    return fig, ax

def plot_energy_scatter_log(energy_matrix, index_to_key):
    """
    Scatter plot with log-scaled energy values.
    """
    # Get log-scaled data
    log_matrix, shift_amount = shift_and_log_data(energy_matrix)

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Extract valid points
    x, y, z = [], [], []
    for i in range(log_matrix.shape[0]):
        for j in range(log_matrix.shape[1]):
            if not np.isnan(log_matrix[i, j]):
                x.append(i)
                y.append(j)
                z.append(log_matrix[i, j])

    scatter = ax.scatter(x, y, z, c=z, cmap='viridis', s=50, alpha=0.6)

    ax.set_xlabel('(Residue, Rotamer) Index - i')
    ax.set_ylabel('(Residue, Rotamer) Index - j')
    ax.set_zlabel('log₁₀(Energy + shift)')
    ax.set_title(f'Protein Energy Landscape - Scatter Plot (Log Scale, shifted by {shift_amount:.2f})')

    fig.colorbar(scatter, ax=ax, label='log₁₀(Energy + shift)')

    return fig, ax
```

```python
from scipy.interpolate import griddata, Rbf
from scipy.ndimage import gaussian_filter

def plot_energy_landscape_triangulated_smooth_griddata(energy_matrix, index_to_key):
    """
    Creates a smoothed 3D surface using griddata interpolation.
    This is fast and generally produces good results.
    """
    # Extract valid (non-NaN) points
    valid_points = []
    valid_energies = []

    for i in range(energy_matrix.shape[0]):
        for j in range(energy_matrix.shape[1]):
            if not np.isnan(energy_matrix[i, j]):
                valid_points.append([i, j])
                valid_energies.append(energy_matrix[i, j])

    valid_points = np.array(valid_points)
    valid_energies = np.array(valid_energies)

    # Create a finer grid for interpolation
    x_min, x_max = valid_points[:, 0].min(), valid_points[:, 0].max()
    y_min, y_max = valid_points[:, 1].min(), valid_points[:, 1].max()

    # Create a denser grid (increase 100 to 200 for even smoother results)
    grid_x = np.linspace(x_min, x_max, 100)
    grid_y = np.linspace(y_min, y_max, 100)
    grid_X, grid_Y = np.meshgrid(grid_x, grid_y)

    # Interpolate using cubic method (smooth) or linear (faster)
    grid_Z = griddata(valid_points, valid_energies, (grid_X, grid_Y), method='cubic')

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    surf = ax.plot_surface(grid_X, grid_Y, grid_Z, cmap='viridis', alpha=0.9,
                           edgecolor='none', antialiased=True)

    ax.set_xlabel('(Residue, Rotamer) Index - j')
    ax.set_ylabel('(Residue, Rotamer) Index - i')
    ax.set_zlabel('Energy')
    ax.set_title('Protein Energy Landscape - Smoothed (GridData Interpolation)')

    fig.colorbar(surf, ax=ax, label='Energy (kcal/mol)')

    return fig, ax

def plot_energy_landscape_triangulated_smooth_rbf(energy_matrix, index_to_key, smooth=0.1):
    """
    Creates a smoothed 3D surface using RBF (Radial Basis Function) interpolation.
    This provides very smooth results and handles sparse data well.

    Args:
        smooth: Smoothing parameter (0 = exact fit, higher = smoother)
    """
    # Extract valid (non-NaN) points
    valid_points = []
    valid_energies = []

    for i in range(energy_matrix.shape[0]):
        for j in range(energy_matrix.shape[1]):
            if not np.isnan(energy_matrix[i, j]):
                valid_points.append([i, j])
                valid_energies.append(energy_matrix[i, j])

    valid_points = np.array(valid_points)
    valid_energies = np.array(valid_energies)

    # Create RBF interpolator
    rbf = Rbf(valid_points[:, 0], valid_points[:, 1], valid_energies,
              function='thin_plate', smooth=smooth)

    # Create a finer grid
    x_min, x_max = valid_points[:, 0].min(), valid_points[:, 0].max()
    y_min, y_max = valid_points[:, 1].min(), valid_points[:, 1].max()

    grid_x = np.linspace(x_min, x_max, 100)
    grid_y = np.linspace(y_min, y_max, 100)
    grid_X, grid_Y = np.meshgrid(grid_x, grid_y)

    # Evaluate RBF on the grid
    grid_Z = rbf(grid_X, grid_Y)

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    surf = ax.plot_surface(grid_X, grid_Y, grid_Z, cmap='viridis', alpha=0.9,
                           edgecolor='none', antialiased=True)

    ax.set_xlabel('(Residue, Rotamer) Index - j')
    ax.set_ylabel('(Residue, Rotamer) Index - i')
    ax.set_zlabel('Energy')
    ax.set_title(f'Protein Energy Landscape - Smoothed (RBF, smooth={smooth})')

    fig.colorbar(surf, ax=ax, label='Energy (kcal/mol)')

    return fig, ax

def plot_energy_landscape_triangulated_smooth_rbf_log(energy_matrix, index_to_key, smooth=0.1):
    """
    Creates a smoothed 3D surface with log scaling using RBF interpolation.
    """
    log_matrix, shift_amount = shift_and_log_data(energy_matrix)

    # Extract valid (non-NaN) points
    valid_points = []
    valid_energies = []

    for i in range(log_matrix.shape[0]):
        for j in range(log_matrix.shape[1]):
            if not np.isnan(log_matrix[i, j]):
                valid_points.append([i, j])
                valid_energies.append(log_matrix[i, j])

    valid_points = np.array(valid_points)
    valid_energies = np.array(valid_energies)

    # Create RBF interpolator
    rbf = Rbf(valid_points[:, 0], valid_points[:, 1], valid_energies,
              function='thin_plate', smooth=smooth)

    # Create finer grid
    x_min, x_max = valid_points[:, 0].min(), valid_points[:, 0].max()
    y_min, y_max = valid_points[:, 1].min(), valid_points[:, 1].max()

    grid_x = np.linspace(x_min, x_max, 100)
    grid_y = np.linspace(y_min, y_max, 100)
    grid_X, grid_Y = np.meshgrid(grid_x, grid_y)

    grid_Z = rbf(grid_X, grid_Y)

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    surf = ax.plot_surface(grid_X, grid_Y, grid_Z, cmap='viridis', alpha=0.9,
                           edgecolor='none', antialiased=True)

    ax.set_xlabel('(Residue, Rotamer) Index - j')
    ax.set_ylabel('(Residue, Rotamer) Index - i')
    ax.set_zlabel('log(Energy + shift)')
    ax.set_title(f'Protein Energy Landscape - Smoothed (RBF, smooth={smooth}, Log Scale)')

    fig.colorbar(surf, ax=ax, label='log(Energy + shift)')

    return fig, ax
```
